```python
import numpy as np
import matplotlib.pyplot as plt

L = 10                  #  
T_values = [1.25, 2.5, 5]  #  
Nx = 100                #    
Delta_x = L / Nx        #   
Delta_t = 0.5 * Delta_x  #    (   )

x = np.linspace(0, L, Nx+1)

u_initial = np.cos(x)**2

def lax_friedrichs(u0, T, Delta_x, Delta_t):
    Nt = int(T / Delta_t)  #   
    u = u0.copy()          #   
    u_new = np.zeros_like(u)  #   
    
    for n in range(Nt):
        for j in range(1, Nx):
            u_new[j] = 0.5 * (u[j] + u[j+1]) - (Delta_t / (2 * Delta_x)) * (u[j+1] - u[j])
        
        u_new[0] = 1  
        u_new[-1] = u[-1]  

        u[:] = u_new[:]
    return u

plt.figure(figsize=(18, 8))

for i, T in enumerate(T_values):
    u = lax_friedrichs(u_initial, T, Delta_x, Delta_t)
    plt.plot(x, u, label=f"T = {T:.2f}")

plt.title("    -")
plt.xlabel("x")
plt.ylabel("u(x, T)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

L = 10                  #  
T_values = [1.25, 2.5, 5]  #   
Nx = 100                #    
Delta_x = L / Nx        #   
Delta_t = 0.5 * Delta_x  

x = np.linspace(0, L, Nx+1)

u_initial = np.cos(x)**2

#  f(U)    (   ,  )
def f(u):
    return u  #     

#  -     F_LF
def lax_friedrichs_with_flux(u0, T, Delta_x, Delta_t, alpha):
    Nt = int(T / Delta_t)  #   
    u = u0.copy()          #   
    u_new = np.zeros_like(u)  #   

    for n in range(Nt):
        for j in range(1, Nx):
            flux = 0.5 * (f(u[j]) + f(u[j+1])) - alpha * (u[j+1] - u[j])
            u_new[j] = u[j] - (Delta_t / Delta_x) * (flux - 0.5 * (f(u[j-1]) + f(u[j]) - alpha * (u[j] - u[j-1])))

        u_new[0] = 1 
        u_new[-1] = u[-1]

        u[:] = u_new[:]

    return u

alpha = 0.5

plt.figure(figsize=(12, 8))

for i, T in enumerate(T_values):
    u = lax_friedrichs_with_flux(u_initial, T, Delta_x, Delta_t, alpha)
    plt.plot(x, u, label=f"T = {T:.2f}")

plt.title("    -   ")
plt.xlabel("x")
plt.ylabel("u(x, T)")
plt.legend()
plt.grid()
plt.show()
```
